#### **Convertir archivos xlsx a csv**

```python
from pathlib import Path
import pandas as pd
```

```python
# Ruta de los archivos fuente
RUTA_DATOS = Path("Datos")
```

Obtener la lista de archivos

```python
archivos = sorted(RUTA_DATOS.glob("*.xlsx"))

print(f"Archivos encontrados: {len(archivos)}")

for archivo in archivos:
    print(f"- {archivo.name}")
```

```python
# Ruta de los archivos CSV
RUTA_DATOS_CSV = Path("Datos CSV")
```

Convertir de xlsx a csv

```python
for archivo in archivos:

    print(f"\nProcesando: {archivo.name}")

    df = pd.read_excel(
        archivo,
        sheet_name="Data"
    )

    nombre_csv = archivo.stem + ".csv"
    ruta_salida = RUTA_DATOS_CSV / nombre_csv

    df.to_csv(
        ruta_salida,
        index=False,
        encoding="utf-8-sig"
    )

    print(f"Guardado: {nombre_csv}")
    print(f"Registros: {len(df):,}")
```

```python
csvs = sorted(RUTA_DATOS_CSV.glob("*.csv"))

print(f"\nCSV generados: {len(csvs)}")

for archivo in csvs:
    print(archivo.name)
```
